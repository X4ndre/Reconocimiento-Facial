```python
import cv2
import numpy as np
```

```python
# Se conecta con el clasificador de caras
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Funcion para la deteccion de caras
def face_extractor(img):
    # La funci√≥n detecta caras y devuelve la cara ajustada
    # Si no se detecta una cara se envia un mensaje de ello
    
    #gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(img, 1.3, 5)
    
    if faces is ():
        return None
    
    # Crop all faces found
    for (x,y,w,h) in faces:
        x=x-10
        y=y-10
        cropped_face = img[y:y+h+50, x:x+w+50]

    return cropped_face
```

```python
# Iniciar Camara
cap = cv2.VideoCapture(0)
count = 300
while True:

    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (400, 400))

        if count < 401:
            file_name_path = './Data/Train/Martin/Martin' + str(count) + '.jpg'
            cv2.imwrite(file_name_path, face)
        elif count >= 401:
            file_name_path = './Data/Test/Martin/Martin' + str(count) + '.jpg'
            cv2.imwrite(file_name_path, face)

        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        print("Face not found")
        pass

    if cv2.waitKey(1) == 13 or count == 500: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Collecting Samples Complete")
```
